```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

df = pd.read_excel('jobs.xlsx')

# שדה הטייטלים - שים כאן את שם העמודה אצלך
job_col = "main_job_title"

# Vectorize using TF-IDF
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(df[job_col].astype(str))

# ---- ELBOW METHOD ----
inertia = []
K = range(1, 15)  # טווח מומלץ לנסיון
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

# ליצור גרף Elbow
plt.figure(figsize=(8,4))
plt.plot(K, inertia, marker='o')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method For Optimal k')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from collections import Counter

df = pd.read_excel('jobs.xlsx')

# שדה הטייטלים - שים כאן את שם העמודה אצלך
job_col = "main_job_title"

# TF-IDF
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(df[job_col].astype(str))

# 1) Reduce the TF-IDF dimensions to 2D
pca = PCA(n_components=2, random_state=42)
X_2d = pca.fit_transform(X.toarray())  # convert sparse → dense for PCA

# KMeans
k = 10 
kmeans = KMeans(n_clusters=k, random_state=42)
clusters = kmeans.fit_predict(X_2d)
df['cluster'] = clusters

# Finding the most common word in each cluster
cluster_labels = {}
for c in range(k):
    titles_in_cluster = df[df['cluster'] == c][job_col].str.lower().str.split()
    all_words = [word for sublist in titles_in_cluster for word in sublist]
    most_common_word = Counter(all_words).most_common(1)[0][0]
    cluster_labels[c] = most_common_word.capitalize()

df['main_job_title'] = df['cluster'].map(cluster_labels)

# Drawing a UMAP with Generic names
plt.figure(figsize=(10, 6))
plt.scatter(X_2d[:, 0], X_2d[:, 1],
            c=clusters, cmap='tab20', s=8, alpha=0.6)

# Plotting centroids with Generic names beside them
for c in range(k):
    x, y = kmeans.cluster_centers_[c]
    plt.text(x, y, cluster_labels[c], fontsize=8, fontweight='bold', color='red')

# 2) Plot
plt.title("PCA + KMeans (Main Job Titles)")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import umap.umap_ as umap
from collections import Counter

# Power BI input table
df = pd.read_excel('jobs.xlsx')

# שדה הטייטלים - שים כאן את שם העמודה אצלך
job_col = "main_job_title"


# TF-IDF text vectorization
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(df[job_col].astype(str))

# ---- UMAP 2D reduction ----
reducer = umap.UMAP(
    n_components=2,
    n_neighbors=20,      # tune for tighter or looser clusters
    min_dist=0.1,        # smaller = more separated clusters
    metric='cosine',
    random_state=42
)

X_2d = reducer.fit_transform(X)

# KMeans
k=10
kmeans = KMeans(n_clusters=k, random_state=42)
clusters = kmeans.fit_predict(X_2d)
df['cluster'] = clusters

# Finding the most common word in each cluster
cluster_labels = {}
for c in range(k):
    titles_in_cluster = df[df['cluster'] == c][job_col].str.lower().str.split()
    all_words = [word for sublist in titles_in_cluster for word in sublist]
    most_common_word = Counter(all_words).most_common(1)[0][0]
    cluster_labels[c] = most_common_word.capitalize()

df['main_job_title'] = df['cluster'].map(cluster_labels)

# Drawing a UMAP with Generic names
plt.figure(figsize=(10, 6))
plt.scatter(X_2d[:, 0], X_2d[:, 1],
            c=clusters, cmap='tab20', s=8, alpha=0.6)

# Plotting centroids with Generic names beside them
for c in range(k):
    x, y = kmeans.cluster_centers_[c]
    plt.text(x, y, cluster_labels[c], fontsize=8, fontweight='bold', color='red')

plt.title("UMAP + KMeans (Main Job Titles)")
plt.xlabel("UMAP 1")
plt.ylabel("UMAP 2")
plt.grid(True)
plt.show()
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Data
df = pd.read_excel('jobs.xlsx')
job_col = "main_job_title"

# 1. Embed job titles semantically
model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(df[job_col].astype(str).tolist(), show_progress_bar=True)

#Reducing dimensions with PCA to 2D (for plotting & clustering)
pca_2d = PCA(n_components=2, random_state=42)
X_2d = pca_2d.fit_transform(embeddings)

# 3. KMeans clustering
k = 10
kmeans = KMeans(n_clusters=k, random_state=42)
clusters = kmeans.fit_predict(embeddings)
df['cluster'] = clusters

# 4. Cluster labeling: closest job title to centroid
cluster_labels = {}
for c in range(k):
    titles_in_cluster = df[df['cluster'] == c][job_col].str.lower().str.split()
    all_words = [word for sublist in titles_in_cluster for word in sublist]
    if all_words:
        most_common_word = Counter(all_words).most_common(1)[0][0]
        cluster_labels[c] = most_common_word.capitalize()
    else:
        cluster_labels[c] = "Unknown"

df['main_job_title'] = df['cluster'].map(cluster_labels)

# Plotting clusters in 2D PCA space
plt.figure(figsize=(10, 6))
plt.scatter(X_2d[:, 0], X_2d[:, 1],
            c=clusters, cmap='tab20', s=8, alpha=0.6)

# Plotting centroids with labels
for c in range(k):
    x, y = X_2d[clusters==c].mean(axis=0)
    plt.text(x, y, cluster_labels[c], fontsize=12, fontweight='bold', color='white')

plt.title("Semantic PCA + KMeans (Main Job Titles)")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.grid(True)
plt.show()
```
